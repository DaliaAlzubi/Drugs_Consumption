```python
drug_type = ['Alcohol', 'Amphet', 'Amyl', 'Benzos', 'Caff', 'Cannabis', 'Choc', 'Coke', 'Crack', 'Ecstasy',
             'Heroin', 'Ketamine', 'Legalh', 'LSD', 'Meth', 'Mushrooms', 'Nicotine', 'VSA']

x = [X_1, X_2, X_3, X_4, X_5, X_6, X_7, X_8, X_9, X_10, X_11, X_12, X_13, X_14, X_15, X_16, X_17, X_18, X_19]
x_train = []
for u in x:
    scaler = MinMaxScaler(feature_range=(0,1))
    uu = scaler.fit_transform(u)
    x_train.append(uu)
y_train = [Y_1, Y_2, Y_3, Y_4, Y_5, Y_6, Y_7, Y_8, Y_9, Y_10, Y_11, Y_12, Y_13, Y_14, Y_15, Y_16, Y_17, Y_18]
```

```python
for i in range(18):
    print(drug_type[i], ":")
    # 2- stochastic gradient descent:
    sgd = SGDClassifier(max_iter=5, tol=None)
    s2 = cross_val_score(sgd, x_train[i], y_train[i], cv=5, scoring = "accuracy")
    print("Stochastic Gradient Descent: ", s2.mean())
    method_names[i].append("SGD")
    method_scores[i].append(s2.mean())
    # 3- Logistic Regression :
    log_reg = LogisticRegression()
    s3 = cross_val_score(log_reg, x_train[i], y_train[i], cv=5, scoring = "accuracy")
    print("Logistic Regression: ", s3.mean())
    method_names[i].append("LR")
    method_scores[i].append(s3.mean())
    # 4- KNN:
    knn = KNeighborsClassifier(n_neighbors=7)
    s4 = cross_val_score(knn, x_train[i], y_train[i], cv=5, scoring = "accuracy")
    print("K Neighbors: ", s4.mean())
    method_names[i].append("KNN")
    method_scores[i].append(s4.mean())
    # 5- Random Forest:
    rand_forest = RandomForestClassifier(n_estimators=100, random_state=42)
    s5 = cross_val_score(rand_forest, x_train[i], y_train[i], cv=5, scoring = "accuracy")
    print("Random Forest: ", s5.mean())
    method_names[i].append("RF")
    method_scores[i].append(s5.mean())
    # 7- Decision Tree:
    dec_tree = DecisionTreeClassifier()
    s7 = cross_val_score(dec_tree, x_train[i], y_train[i], cv=5, scoring = "accuracy")
    print("Decision Tree: ", s7.mean())
    method_names[i].append("DT")
    method_scores[i].append(s7.mean())
    print("\n\n")
```
